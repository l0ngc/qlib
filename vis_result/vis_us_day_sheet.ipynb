```python
import os
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from pprint import pprint

# rsync -avz sx:/16t-2/cl/project/qlib/vis_result/plots/ /Users/chenglong/Documents/plots
# Constants for file names and model paths

PERFORMANCE_FILE = 'port_analysis_1day.pkl'
REPORT_FILE = 'report_normal_1day.pkl'
DIR_AFTERFIX = 'portfolio_analysis'

# Result directories
RESULT_PATH = '/16t-2/cl/project/qlib/examples/us_day_res'

RR_PATH = '/home/cl/project/qlib/examples/rr_result_collected'
DDGDA_PATH = '/home/cl/project/qlib/examples/ddgda_result_collected'

TRAINING_TYPES = ['ST', 'RR', 'DDGDA']
MODEL_LIST = ['LightGBM', 'Transformer', 'Linear', 'LSTM', 'XGBoost', 'MLP']
PERF_TYPE = 'return_without_cost'
# MODEL_LIST = ['LightGBM', 'Transformer', 'Linear', 'LSTM', 'DoubleEnsemble', 'XGBoost', 'CatBoost', 'MLP']
```

```python
# scan done models
perf_dict = {}
for tt in TRAINING_TYPES:
    tmp_dict = {}
    for md in MODEL_LIST:
        try:
            file_path = os.path.join(RESULT_PATH, tt, md, PERFORMANCE_FILE)
            perf_res = pd.read_pickle(file_path)
            tmp_dict[md] = perf_res.T
        except:
            print(f'Training Type {tt}, model: {md} is missed')
            continue
    perf_dict[tt] = tmp_dict
```

```python
pd.concat(perf_dict['RR'])
```

```python
pd.concat(perf_dict['DDGDA'])
```

```python
# Adding a new column to each dataframe to indicate the method
df_ST = pd.concat(perf_dict['ST'])
df_RR = pd.concat(perf_dict['RR'])
df_DDGDA = pd.concat(perf_dict['DDGDA'])

df_ST = df_ST[PERF_TYPE]
df_RR = df_RR[PERF_TYPE]
df_DDGDA = df_DDGDA[PERF_TYPE]

df_ST['Method'] = 'ST'
df_RR['Method'] = 'RR'
df_DDGDA['Method'] = 'DDGDA'
```

```python
combined_df = pd.concat([df_ST, df_RR, df_DDGDA])
combined_df = combined_df.reset_index()
combined_df.columns = ['Model', 'drop', 'mean', 'std', 'annualized_return', 'information_ratio', 'max_drawdown', 'Method']
combined_df = combined_df[['Model', 'mean', 'std', 'annualized_return', 'information_ratio', 'max_drawdown', 'Method']]
```

```python
# Defining the order for Model and Method
model_order = ['Linear', 'MLP', 'LightGBM', 'CatBoost', 'LSTM', 'Transformer', 'XGBoost']
method_order = ['ST', 'RR', 'DDGDA']

# Converting to categorical to enforce the order
combined_df['Model'] = pd.Categorical(combined_df['Model'], categories=model_order, ordered=True)
combined_df['Method'] = pd.Categorical(combined_df['Method'], categories=method_order, ordered=True)

# Sorting the dataframe by Model and Method
combined_df.sort_values(by=['Model', 'Method'], inplace=True)

# Reordering columns for better readability
combined_df = combined_df[['Model', 'Method', 'mean', 'std', 'annualized_return', 'information_ratio', 'max_drawdown']]

# Setting multi-index for proper formatting
combined_df.set_index(['Model', 'Method'], inplace=True)
```
